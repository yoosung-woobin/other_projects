```python
from collections import deque
import numpy as np
import copy
```

```python
state = ((2,8,3),
         (1,6,4),
         (7,None,5))
goal_state =((1,2,3),
            (8,None,4),
            (7,6,5))
```

```python
def possible_node(node) :
    n = []
    for item in [(1,0),(0,1),(-1,0),(0,-1)] :
        next_x = node[0]+item[0]
        next_y = node[1]+item[1]
        if next_x<0 or next_x>2 or next_y<0 or next_y>2:
            continue
        n.append((next_x,next_y))
    return n
```

```python
def switch_value(array,ori,new):
    #"ori" denotes the index of given node (takes form of tuple)
    #"new" denotes the index of target node(takes form of tuple)
    #This function takes input of array and returns modified array after two nodes are exchanged
    subj = copy.deepcopy(array)
    subj = [list(item) for item in subj]
    switcher = subj[ori[0]][ori[1]]
    switchee = subj[new[0]][new[1]]
    subj[ori[0]][ori[1]] = switchee
    subj[new[0]][new[1]] = switcher
    subj = tuple(tuple(item) for item in subj)
    return subj
```

```python
def square_locator(state):
    square = [(index,state[index].index(None)) for index,item in enumerate(state) if None in state[index] ].pop()
    return square
```
